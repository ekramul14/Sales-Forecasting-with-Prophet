# TRAIN THE MODEL PART B


   - StateHoliday: indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
   - SchoolHoliday: indicates if the (Store, Date) was affected by the closure of public schools
  

```python
sales_train_all_df.head()
```

```python
def sales_predictions(Store_id, sales_df,holidays, periods):
  sales_df = sales_df[sales_df['Store'] == Store_id]
  sales_df = sales_df[['Date', 'Sales']].rename(columns={'Date': 'ds', 'Sales': 'y'})
  sales_df.sort_values('ds')

  model = Prophet(holidays = holidays)
  model.fit(sales_df)
  future = model.make_future_dataframe(periods = periods)
  forecast = model.predict(future)
  figure = model.plot(forecast, xlabel = 'Date', ylabel = 'Sales')
  figure2 = model.plot_components(forecast)

```

```python
# Get all the dates pertaining to school holidays
school_holidays = sales_train_all_df[sales_train_all_df['SchoolHoliday'] == 1].loc[:, 'Date'].values
```

```python
school_holidays
```

```python
# Get all the dates pertaining to state holidays
state_holidays = sales_train_all_df[(sales_train_all_df['StateHoliday'] == 'a') | (sales_train_all_df['StateHoliday'] == 'b') | (sales_train_all_df['StateHoliday'] == 'c')].loc[:, 'Date'].values
```

```python
from sklearn.metrics import mean_absolute_error

# Modified sales_predictions function to include returning forecast and actual data
def sales_predictions_with_mape(Store_id, sales_df, holidays, periods):
    sales_df = sales_df[sales_df['Store'] == Store_id]
    sales_df = sales_df[['Date', 'Sales']].rename(columns={'Date': 'ds', 'Sales': 'y'})
    sales_df.sort_values('ds', inplace=True)

    # Initialize and train the Prophet model
    model = Prophet(holidays=holidays)
    model.fit(sales_df)

    # Make predictions
    future = model.make_future_dataframe(periods=periods)
    forecast = model.predict(future)

    # Extract actual and predicted sales for the test period
    actual = sales_df['y'].values  # Actual sales
    predicted = forecast['yhat'].iloc[:len(actual)].values  # Predicted sales for the same period

    # Avoid division by zero in MAPE calculation
    non_zero_indices = actual != 0
    actual = actual[non_zero_indices]
    predicted = predicted[non_zero_indices]

    # Calculate MAPE
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

    # Return the forecast and plots for further analysis
    model.plot(forecast, xlabel='Date', ylabel='Sales')
    model.plot_components(forecast)
    return forecast

# Example: Calculate MAPE for a specific store with holidays considered
forecast = sales_predictions_with_mape(6, sales_train_all_df, school_state_holidays, 90)
```
